```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
from deep_autoviml import deep_autoviml as deepauto
```

```python
#Uploaded Kaggle dataset and unzip
!unzip archive.zip
```

```python
train = pd.read_csv('aug_train.csv')
train.head()
```

```python
target = 'target'
```

```python
train, test = train_test_split(train, test_size=0.2, random_state=99)
test.head()
```

```python
keras_model_type =  'fast1'
model_options = {"tuner":"optuna", 'nlp_char_limit':10, 'cat_feat_cross_flag':False, 'max_trials': 5}
keras_options = {"patience":10,'class_weight':True, "early_stopping": True, "epochs": 100}
```

```python
#One line for AutoViML to create and train the model 
output = deepauto.fit(train, target, keras_model_type,project_name='tabular_project', keras_options=keras_options,
                 model_options=model_options, save_model_flag=False, use_my_model='', verbose=1)
```

```python
model = output[0]
cat_vocab_dict = output[1]
```

```python
#Predictions using the 20% test data from earlier split
y_preds = deepauto.predict(model, 'tabular_project', test_dataset=test,
                                 keras_model_type=keras_model_type, cat_vocab_dict=cat_vocab_dict)
```

```python
y_test = test[target].values
```
